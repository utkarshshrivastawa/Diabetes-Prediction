# Diabetes Prediction Using Support Vector Machine

```python
# importing the library
import numpy as np 
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

##  Data Standardization

```python
Scaler = StandardScaler()
```

```python
Scaler.fit(x)
```

```python
standardized_data =Scaler.transform(x) # we can also do Scaler.fit(x).transform(x)
```

```python
standardized_data
```

```python
x = standardized_data
y = data['Outcome']
```

## Train Test and Split

```python
X_train,X_test,Y_train ,Y_test = train_test_split(x,y,test_size=0.2 , stratify = y ,random_state = 2)
```

```python
# count the number of rows and columns 
print(x.shape , X_train.shape , X_test.shape)
```

```python
print(X_train)
```

```python
print(X_train)
```

## Training the model 

```python
classifier = svm.SVC(kernel = 'linear')
```

```python
# training the Support Vector Machine Classifier 
classifier.fit(X_train , Y_train)
```

## Model Evaluation

```python
# accuracy score for training data 
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
print("The accuracy of train data is : ",training_data_accuracy)
```

```python
# accuracy score for test data 
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
print("The accuracy of test data is :" , test_data_accuracy)
```

## Making a Predictive System

```python
# input_data = (6,148,72,35,0,33.6,0.627,50)
input_data = (1,89,66,23,94,28.1,0.167,21)
# changing the input data to numpy array 
input_data_as_array = np.asarray(input_data)

# reshape the array as we are predicting  for one instance 
input_data_reshaped = input_data_as_array.reshape(1,-1)

# standardize the input dat 
std_data = Scaler.transform(input_data_reshaped )
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction == 1):
    print("The person is diabetes")
else :
    print("The person is non-diabetes")
```
